```python
from ultralytics import YOLO
model = YOLO("yolov8s.pt")  # Load a pre-trained YOLO v8 small model
```

```python
model.info()
```

```python
print(model)
```

```python
freeze_layers = 40
for idx, (name, param) in enumerate(model.named_parameters()):
    if idx < freeze_layers:  # Freeze first 10 layers
        param.requires_grad = False

```

```python
model.train(
    data="/content/Zebra-Crossing-1/data.yaml",
    epochs=100,
    batch=16,   # Adjust batch size based on GPU memory
)
```

```python
model.val()  # Evaluate on validation set
```
